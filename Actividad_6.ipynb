```python
def mcd(a, b):
    if a == b:
        return a
    elif a > b:
        return mcd(a - b, b)
    else:
        return mcd(a, b - a)

# Solicitud al cliente:
a = int(input("Cliente, ingrese el primer número: "))
b = int(input("Cliente, ingrese el segundo número: "))
resultado_mcd = mcd(a, b)
print("El máximo común divisor es:", resultado_mcd)
```

```python
def invertir_cadena(cadena):
    if len(cadena) == 0:
        return cadena
    else:
        return invertir_cadena(cadena[1:]) + cadena[0]

# Solicitud al cliente:
cadena = input("Cliente, ingrese una cadena: ")
cadena_invertida = invertir_cadena(cadena)
print("La cadena invertida es:", cadena_invertida)
```

```python
def maximo_vector(vector):
    if len(vector) == 1:
        return vector[0]
    else:
        max_rest = maximo_vector(vector[1:])
        return vector[0] if vector[0] > max_rest else max_rest

# Solicitud al cliente:
n = int(input("Cliente, ingrese el tamaño del vector: "))
vector = [int(input(f"Ingrese el valor {i+1}: ")) for i in range(n)]
maximo = maximo_vector(vector)
print("El valor máximo es:", maximo)
```

```python
def suma_vector(vector):
    if len(vector) == 0:
        return 0
    else:
        return vector[0] + suma_vector(vector[1:])

# Solicitud al cliente:
n = int(input("Cliente, ingrese el tamaño del vector: "))
vector = [int(input(f"Ingrese el valor {i+1}: ")) for i in range(n)]
suma_total = suma_vector(vector)
print("La suma de los elementos es:", suma_total)
```

```python
def valor_igual_indice(vector, index=0):
    if index == len(vector):
        return False
    if vector[index] == index:
        return True
    return valor_igual_indice(vector, index + 1)

# Solicitud al cliente:
n = int(input("Cliente, ingrese el tamaño del vector: "))
vector = [int(input(f"Ingrese el valor {i+1}: ")) for i in range(n)]
coincide = valor_igual_indice(vector)
print("¿Algún valor coincide con su índice?:", coincide)
```
