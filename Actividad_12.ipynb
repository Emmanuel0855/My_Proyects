```python
def bubble_sort(arreglo):
    n = len(arreglo)
    for i in range(n, 0, -1):
        for j in range(i - 1):
            if arreglo[j] > arreglo[j + 1]:
            #Intercambiar elementos
                aux = arreglo[j]
                arreglo[j] = arreglo[j + 1]
                arreglo[j + 1] = aux

arreglo = [5, 2, 9, 1, 5, 6]
bubble_sort(arreglo)
print(arreglo)
```

```python
def shake_sort(array):
    n = len(array)
    l = 1
    r = n-1
    k = n-1
    while l<r:
        j = r
        while j>0:
            if array[j-1] > array[j]:
                aux = array[j-1]
                array[j-1] = array[j]
                array[j] = aux
                k = j
            j -=1
        l = k + 1


array = [5,2,9,1,5,6]
shake_sort(array)
print(array)
```

```python
def insertion_sort(a): 
    n = len(a) 
    for i in range(1, n): 
        j=1-1 
        t = a[i] 
        while j >= 0 and t < a[j]:
            a[j + 1] = a[j] 
            j -= 1
        a[j + 1] = t

# Ejemplo de uso 
arreglo = [5, 2, 9, 1, 5, 6] 
insertion_sort(arreglo) 
print(arreglo)
```

```python
def selection_sort(array):
    i = len(array)
    while i > 0:
        for j in range(i-1):
            if array[j] > array[j + 1]:
                aux = array[j]
                array[j] = array[j + 1]
                array[j + 1] = aux
        i -=1

array = [5, 2, 9, 1, 5, 6]
selection_sort(array)
print(array)
```

```python
def merge_sort(array):
    size = len(array) -1
    for i in range(0, size):
        j = i
        minor = array[j]
        for k in range(i, size):
            siguiente = array[k+1]
            if minor > siguiente:
                minor = siguiente
                j=k+1
        if j !=i:
            minor = array[j]
            array[j]=array[i]
            array[i] = minor

array = [5, 2, 9, 1, 5, 6]
merge_sort(array)
print(array)
```
