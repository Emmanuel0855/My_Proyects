```python
def contar_vocales(cadena):
    vocales = "aeiouAEIOU"
    contador = 0
    for letra in cadena:
        if letra in vocales:
            contador += 1
    return contador

# Input de la terminal
cadena = input("Ejercicio 1 - Ingresa una cadena de texto: ")
print(f"Número de vocales: {contar_vocales(cadena)}")
```

```python
def factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n - 1)

# Input de la terminal
numero = int(input("Ejercicio 2 - Ingresa un número para calcular su factorial: "))
print(f"El factorial de {numero} es: {factorial(numero)}")
```

```python
def es_palindromo(palabra):
    palabra = palabra.lower().replace(" ", "")  # Ignorar mayúsculas y espacios
    return palabra == palabra[::-1]

# Input de la terminal
palabra = input("Ejercicio 3 - Ingresa una palabra o frase para verificar si es palíndromo: ")
print(f"¿Es palíndromo?: {es_palindromo(palabra)}")
```

```python
def es_primo(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

def numeros_primos(hasta):
    primos = []
    for num in range(2, hasta + 1):
        if es_primo(num):
            primos.append(num)
    return primos

# Input de la terminal
limite = int(input("Ejercicio 4 - Ingresa un número límite para generar los números primos: "))
print(f"Números primos hasta {limite}: {numeros_primos(limite)}")
```
