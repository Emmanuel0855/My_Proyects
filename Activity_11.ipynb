```python
def invertir(cadena):
    pila = []
    # Apilamos cada carácter de la cadena
    for char in cadena:
        pila.append(char)
    
    # Desapilamos para invertir la cadena
    invertida = ''
    while pila:
        invertida += pila.pop()
    
    return invertida

# Ejemplo
print(invertir("abcd"))  # Salida: "dcba"
```

```python
def binario(numero):
    pila = []
    
    # Convertimos a binario usando la pila
    while numero > 0:
        pila.append(numero % 2)
        numero //= 2
    
    # Desapilamos para obtener el número binario
    binario = ''
    while pila:
        binario += str(pila.pop())
    
    return binario

# Ejemplo
print(binario(10))  # Salida: "1010"
```

```python
def separarParImpar(pila):
    pares = []
    impares = []
    
    # Separamos los números
    while pila:
        numero = pila.pop()
        if numero % 2 == 0:
            pares.append(numero)
        else:
            impares.append(numero)
    
    # Reconstruimos la pila con los pares abajo y los impares arriba
    while pares:
        pila.append(pares.pop())
    while impares:
        pila.append(impares.pop())
    
    return pila

# Ejemplo
pila = [2, 3, 6, 8, 11, 13, 18, 21]
print(separarParImpar(pila))  # Salida: [2, 6, 8, 18, 3, 11, 13, 21]
```

```python
def sort(pila):
    pila_aux = []
    
    while pila:
        tmp = pila.pop()
        while pila_aux and pila_aux[-1] > tmp:
            pila.append(pila_aux.pop())
        pila_aux.append(tmp)
    
    # Reconstruimos la pila original
    while pila_aux:
        pila.append(pila_aux.pop())
    
    return pila

# Ejemplo
pila = [1, 3, 2, 4]
print(sort(pila))  # Salida: [4, 3, 2, 1]
```

```python
def esValida(secuencia):
    pila = []
    
    for char in secuencia:
        if char == '{':
            pila.append(char)
        elif char == '}':
            if not pila or pila.pop() != '{':
                return False
    
    return len(pila) == 0

# Ejemplo
print(esValida("{{}{{}{}}}"))  # Salida: True
print(esValida("{{}{}{}"))  # Salida: False
```
